5. 损失函数
       RNN网络预测的结果是通过softmax计算概率最高的词做为预算结果, 本质上还是分类, 所以RNN网络的损失函数就是交叉熵损失
   

```python
# concatenate之后的结果, 和分别计算的结果是一样的. 
import torch
```

```python
X, W_xh = torch.normal(0, 1, (3, 1)), torch.normal(0, 1, (1, 4)) # (3, 1) . (1, 4) -> (3, 4)
H, W_hh = torch.normal(0, 1, (3, 4)), torch.normal(0, 1, (4, 4)) # (3, 4) .(4, 4) -> (3, 4)
```

```python
# 分别点乘然后相加
torch.matmul(X, W_xh) + torch.matmul(H, W_hh)
```

```python
# concatenate之后再做点乘.
torch.cat((X, H), dim=1).shape
```

```python
torch.cat((W_xh, W_hh), dim=0).shape
```

```python
torch.matmul(torch.cat((X, H), dim=1), torch.cat((W_xh, W_hh), dim=0))
```
